### Match Statement


The match statement selects one of many code blocks to be executed




```python
x = 2

match x:
    case 1:
        print("one")
    case 2:
        print("two")
    case 3:
        print("three")
```

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

## 2.5 Loop Statement 

For  -> the cound of times          


```python
range(10) # End = 10  
print(range(10))
print(type(range(10)))
```

```python
# for loop 

for i in range(10):
    print(i , end=" - ")
    
print ()

# range (start , end , step) 

for i in range(1 , 10 , 2):
    print(i , end=" - ")
    
print ()

for i in range(10 , 0 , -2):
    print(i , end=" - ")
    
print ()
```

```python
for i in range (5):    # nested loop
    for j in range(5):  
        print(j)

    print(f"Circle {i} Done ")    
```

```python
for i in range (10):   
    for j in range(10):
        if (j < i):
           print("*", end= "") 
    print()  
```

```python
n = 10

for i in range(n):
    print(" " * (n - i - 1) + "*" * (2*i + 1))

for i in range(n-2, -1, -1):
    print(" " * (n - i - 1) + "*" * (2*i + 1))
```
